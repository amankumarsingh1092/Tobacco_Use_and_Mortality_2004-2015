# Conclusion: Tobacco-Related Mortality Analysis (20042015)
This project analyzes tobacco-related and general mortality trends in the UK from 2004 to 2015 using data from NHS admissions and fatalities.

## Key Insights:
* Consistently High Mortality: Total annual deaths remained significantly high with no major decline over the years.
* Smoking's Heavy Toll:
* A large portion of deaths stem from smoking-related illnesses, including lung cancer,          respiratory diseases, and oral cancers.
* Pie chart analysis revealed approximately 3545% of total deaths were smoking-related.
* Gender Disparity: Males showed notably higher fatality rates for tobacco-linked conditions.
* Top Diagnoses by Deaths:
  Lung cancer
  Chronic obstructive pulmonary disease (COPD)
  Ischemic heart diseases
* Heatmaps & Trends revealed spikes in certain diagnoses and years, warranting closer public health attention.

## Visual Tools Used:
* Line plots for year-wise mortality trends
* Bar charts for cause-specific fatalities
* Gender trend comparison
* Heatmaps for ICD10 cause-year interactions
* Pie chart for smoking vs non-smoking death proportions

## Conclusion:
The data highlights tobacco use as a persistent and preventable public health crisis. Despite awareness efforts, its long-term health consequences remain severe. This project emphasizes the need for continued policy action, public education, and smoking cessation programs.
