```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load and preprocess
df = prescriptions
df['Year'] = df['Year'].str.extract(r'(\d{4})').astype(int)

# Plot
plt.figure(figsize=(10,6))
sns.lineplot(data=df, x='Year', y='All Pharmacotherapy Prescriptions', marker='o')
plt.title("Total Tobacco-Related Prescriptions Over Time")
plt.xlabel("Year")
plt.ylabel("Total Prescriptions")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Prescriptions by Drug Type
plt.figure(figsize=(12,6))
sns.lineplot(x='Year', y='Nicotine Replacement Therapy (NRT) Prescriptions', data=df, label='NRT')
sns.lineplot(x='Year', y='Bupropion (Zyban) Prescriptions', data=df, label='Bupropion')
sns.lineplot(x='Year', y='Varenicline (Champix) Prescriptions', data=df, label='Varenicline')
plt.title("Prescriptions by Drug Type Over Time")
plt.ylabel("Number of Prescriptions")
plt.xlabel("Year")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Cost Trends for Each Drug
plt.figure(figsize=(12,6))
sns.lineplot(x='Year', y='Net Ingredient Cost of Nicotine Replacement Therapies (NRT)', data=df, label='NRT')
sns.lineplot(x='Year', y='Net Ingredient Cost of Bupropion (Zyban)', data=df, label='Bupropion')
sns.lineplot(x='Year', y='Net Ingredient Cost of Varenicline (Champix)', data=df, label='Varenicline')
plt.title("Net Cost of Tobacco Cessation Drugs Over Time")
plt.ylabel("Cost (£)")
plt.xlabel("Year")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Pie Chart: Proportion of Prescriptions by Drug (Latest Year)
latest = df[df['Year'] == df['Year'].max()]
labels = ['NRT', 'Bupropion', 'Varenicline']
sizes = [latest['Nicotine Replacement Therapy (NRT) Prescriptions'].values[0],
         latest['Bupropion (Zyban) Prescriptions'].values[0],
         latest['Varenicline (Champix) Prescriptions'].values[0]]

plt.figure(figsize=(7, 7))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['orange', 'blue', 'green'])
plt.title(f"Prescription Share by Drug Type in {latest['Year'].values[0]}")
plt.tight_layout()
plt.show()
```

# Summary – Tobacco Prescription Data Analysis (2004–2015)
## Objective
To analyze trends in tobacco cessation prescriptions and associated costs in the UK from 2004 to 2015, focusing on three major pharmacotherapies:
* Nicotine Replacement Therapy (NRT)
* Bupropion (Zyban)
* Varenicline (Champix)

## Dataset Overview
The dataset includes:
* Number of prescriptions by year and drug type
* Net ingredient cost per pharmacotherapy
* Total prescriptions and total cost per year

## Key Insights
* Peak Years: Total prescriptions for tobacco cessation peaked in 2010–2012.
* Dominant Therapy: NRT was the most commonly prescribed treatment throughout the period.
* Varenicline Surge: Varenicline prescriptions rose sharply until 2012 but declined afterward.
* Cost Trends: Cost patterns closely followed prescription trends; NRT incurred the highest   cumulative cost.
* Bupropion Decline: Use of Bupropion consistently declined after 2008.

## Visualizations Included
* Total prescriptions over time
* Drug-wise prescription trends
* Net ingredient cost trends by drug
* Pie chart showing prescription distribution by drug
* Use of the 'magma' palette for improved visualization and clarity

## Conclusion
The data reflects changing public health preferences and possible shifts in clinical guidelines. A marked shift in prescribing patterns suggests that healthcare strategies, policy decisions, and patient adherence have evolved during this time. This analysis offers valuable insights into cost-effectiveness and utilization trends of pharmacotherapies aimed at tobacco cessation.
